# Question 3 : prédiction par lissage exponentiel 
Séparer vos données en ensemble d’apprentissage et de test pour effectuer des prédictions par lissage 
exponentiel sur les données DPAE (CDD et CDI) et EESSA (secteurs sélectionnés).

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

# Pour le cas des CDI et du CDD

On sépare le jeu de donnée en 70% de jeu de train et 30% de jeu de test. On découpe dans l'ordre car le but final est de réaliser des prédictions sur les dates à venir.

```python
taille_train = int(len(CDD) * 0.7) #CDD et CDI ont la même taille
train_CDI, test_CDI = CDI[:taille_train], CDI[taille_train:]
train_CDD, test_CDD = CDD[:taille_train], CDD[taille_train:]
```

On entrâine le modèle et on réalise la prédiction

```python
model_CDI = ExponentialSmoothing(train_CDI['DPAE (brut)'], trend='add', seasonal='add', seasonal_periods=12) #le modele est additif et non multiplicatif
model_CDI = model_CDI.fit()
pred_model_CDI = model_CDI.forecast(len(test_CDI))
```

```python
plt.figure(figsize = (13,7))
pred_model_CDI.plot() #on trace les prédictions du modèle
```

Idem pour les CDD

```python
plt.figure(figsize = (13,7))
model_CDD = ExponentialSmoothing(train_CDD['DPAE (brut)'], trend='add', seasonal='add', seasonal_periods=12)
model_CDD = model_CDD.fit()
pred_model_CDD = model_CDD.forecast(len(test_CDD))
pred_model_CDD.plot()
```

# Pour le cas des deux secteurs recherche et enseignement

```python
taille_train_recherche = int(len(df_RetD) * 0.7) 
train_recherche, test_recherche = df_RetD[:taille_train_recherche], df_RetD[taille_train_recherche:]
taille_train_enseignement = int(len(df_enseignement) * 0.7) 
train_enseignement, test_enseignement = df_enseignement[:taille_train_enseignement], df_enseignement[taille_train_enseignement:]
```

pour la recherche

```python
plt.figure(figsize = (13,7))
model_recherche = ExponentialSmoothing(train_recherche['Nombre'], trend='add', seasonal='add', seasonal_periods=12)
model_recherche = model_recherche.fit()
pred_model_recherche = model_recherche.forecast(len(test_recherche))
pred_model_recherche.plot()
```

pour l'enseignement

```python
plt.figure(figsize = (13,7))
model_enseignement = ExponentialSmoothing(train_enseignement['Nombre'], trend='add', seasonal='add', seasonal_periods=12)
model_enseignement = model_enseignement.fit()
pred_model_enseignement = model_enseignement.forecast(len(test_enseignement))
pred_model_enseignement.plot()
```

#Question 5 : application et évaluation d’autres modèles de prédiction
Utiliser d’autres modèles (par exemple ARMA) pour effectuer la prédiction. Comparer les résultats.

Modèle 2 : ARIMA (autoregressive integrated moving average)

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
plt.figure(figsize = (13,7))
model2_CDD = ARIMA(train_CDD['DPAE (brut)'], order=(12, 1, 12)) #test de plusieurs order
model2_CDD = model2_CDD.fit()
pred_model2_CDD = model2_CDD.forecast(len(test_CDD))

pred_model2_CDD.plot()
mae_ARIMA = mean_absolute_error(test_CDD['DPAE (brut)'], pred_model2_CDD)
mse_ARIMA = mean_squared_error(test_CDD['DPAE (brut)'], pred_model2_CDD)
rmse_ARIMA = np.sqrt(mse_ARIMA)

# Afficher les mesures d'erreur
print(mae_ARIMA, mse_ARIMA, rmse_ARIMA)
```

Modèle 3 : SARIMA (seasonal autoregressive integrated moving average)

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
plt.figure(figsize = (13,7))
model_SARIMA = SARIMAX(train_CDD['DPAE (brut)'], order=(1,1,1), seasonal_order=(1,1,1,12))
model_SARIMA = model_SARIMA.fit()
pred_SARIMA = model_SARIMA.forecast(len(test_CDD))

pred_SARIMA.plot()
mae_SARIMA = mean_absolute_error(test_CDD['DPAE (brut)'], pred_SARIMA)
mse_SARIMA = mean_squared_error(test_CDD['DPAE (brut)'], pred_SARIMA)
rmse_SARIMA = np.sqrt(mse_SARIMA)

print(mae_SARIMA, mse_SARIMA, rmse_SARIMA)
```
